# Compare Agent Configurations

Configure multiple backends, run them on the same benchmark, and compare pass rates.

```python
from sql_agent.config import Settings
from sql_agent.evaluation import ComparisonSuite
```

```python
shared = dict(
    database_url='sqlite:///../examples/chinook/data/chinook.sqlite',
    allowed_tables=['Album', 'Artist', 'Customer', 'Employee', 'Genre', 'Invoice', 'InvoiceLine', 'MediaType', 'Playlist', 'PlaylistTrack', 'Track'],
    max_rows=200,
    markdown_docs_dir='../examples/chinook/docs/markdown',
    sql_dialect='sqlite',
    database_label='Chinook SQLite database',
    openai_api_key=None,
    openai_base_url=None,
    hf_token='YOUR_HF_TOKEN',
    ollama_base_url='http://localhost:11434/v1',
    google_api_key=None,
    google_api_base=None,
)

gemma4_local = Settings(model_provider='ollama', model_id='gemma4:e2b', **shared)
qwen3_hf = Settings(model_provider='huggingface', model_id='Qwen/Qwen3-4B-Instruct', **shared)
```

```python
suite = ComparisonSuite(dataset='../chinook_benchmark_v1.json.rtf', output_dir='../eval_results')
suite.add_config('gemma4-local', gemma4_local)
suite.add_config('qwen3-hf', qwen3_hf)
results = suite.run_all()
```

```python
comparison_df = suite.compare(results)
comparison_df
```

```python
import matplotlib.pyplot as plt

ax = comparison_df.plot(x='config', y='pass_rate', kind='bar', legend=False, title='Pass rate by configuration')
ax.set_ylabel('Pass rate')
plt.show()
```
