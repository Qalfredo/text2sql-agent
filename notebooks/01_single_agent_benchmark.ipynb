# Single Agent Benchmark

Run one configured agent against the Chinook benchmark and inspect the results interactively.

```python
from dotenv import load_dotenv
from sql_agent.config import load_settings
from sql_agent.evaluation import BenchmarkRunner

load_dotenv('../.env', override=True)
settings = load_settings()
```

```python
runner = BenchmarkRunner(
    agent_config=settings,
    dataset='../chinook_benchmark_v1.json.rtf',
    output_dir='../eval_results',
    max_workers=1,
    timeout_per_question=60,
)
result = runner.run()
result.summary
```

```python
df = result.to_dataframe()
df[['id', 'difficulty', 'pass', 'duration_ms', 'final_sql']].head()
```

```python
import matplotlib.pyplot as plt

ax = df['duration_ms'].plot(kind='hist', bins=12, title='Latency distribution (ms)')
ax.set_xlabel('Duration (ms)')
plt.show()
```
